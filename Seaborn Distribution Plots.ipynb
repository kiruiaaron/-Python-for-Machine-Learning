```python
import seaborn as sns
%matplotlib inline
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
sns.distplot(tip['total_bill'])
```

```python
sns.distplot(tip['total_bill'], kde = False, bins = 30)
```

```python
sns.jointplot(x='total_bill', y='tip', data = tip, kind = 'reg')
```
