```python
import seaborn as sns
%matplotlib inline
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
sns.countplot(x = 'time', data = tip)
```

```python
sns.barplot(x = 'time', y = 'total_bill', data = tip)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data =  tip)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tip, hue = 'time')
```

```python
sns.swarmplot(x = 'time', y = 'total_bill', data = tip)
```

```python
sns.boxplot(x='day', y='total_bill', data = tip)
```

```python
sns.boxplot(x='day', y='total_bill', hue = 'smoker', data = tip)
```
