```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(10)
x
```

```python
y = x*2
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Title')
plt.show()
```

```python
plot=plt.figure()
axplot = plot.add_axes([0,0,1,1])
axplot.plot(x,y)
axplot.set_xlabel('x axis')
axplot.set_ylabel('y axis')
axplot.set_title('Title')

axplot2 = plot.add_axes([0.2,0.2,.5,.5])
axplot2.plot(x,y)
axplot2.set_xlabel('x axis 2')
axplot2.set_ylabel('y axis 2')
axplot2.set_title('Title 2')

```
