```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x =  np.arange(10)
```

```python
y = x* 2
```

```python
x
```

```python
y
```

```python
plot = plt.figure()
axplot = plot.add_axes([0,0,1,1])

axplot.plot(x,y)
axplot.set_xlabel('x axis')
axplot.set_ylabel('y axis')
axplot.set_title('Title')
plt.show()
```

```python
plot = plt.figure()
axplot = plot.add_axes([0,0,1,1])

axplot.plot(x,y, label ='x')
axplot.plot(x*3,y, label= 'x*3')
axplot.legend()
axplot.set_xlabel('x axis')
axplot.set_ylabel('y axis')
axplot.set_title('Title')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color ='blue')
ax.plot(x,x*1.5,color ='blue')
ax.plot(x,x*2,color = 'blue')
ax.plot(x,x*2.5,color = '#B31ED4',a lpha = 0.3)
ax.plot(x,x*3,color = 'blue',linewidth = 1)
ax.plot(x,x*3.5,color = 'blue',lw = 2)
ax.plot(x,x*4,color = 'blue',lw = 3, linestyle = '-.')
ax.plot(x,x*4.5,color = 'blue',lw = 3, ls = ':')
ax.plot(x,x*5,color = 'blue',lw = 1, marker = 's',markersize = 10, markerfacecolor = 'green')
```
