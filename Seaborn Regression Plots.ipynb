```python
import seaborn as sns
%matplotlib inline
```

```python
tip = sns.load_dataset('tips')
```

```python
tip.head()
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data=tip)
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data=tip, hue = 'time')
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data=tip, hue= 'time', markers =['o','s'])
```

```python
 sns.lmplot(x = 'total_bill', y ='tip', data=tip, col ='time')
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data=tip, col='time', size = 1)
```
